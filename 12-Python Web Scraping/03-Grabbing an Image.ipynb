```python
import requests 
```

```python
import bs4
```

Making a __request__

```python
res = requests.get('https://en.wikipedia.org/wiki/Julius_Caesar')
```

Turning it into a __soup__

```python
soup = bs4.BeautifulSoup(res.text,'lxml')
```

```python
soup
```

__Selecting__ a particular __tag__/__class__ call

```python
soup.select('img')
```

Specifying

```python
soup.select('.thumbimage')
```

```python
painting = soup.select('.thumbimage')[4]
```

```python
type(painting)
```

Confirming the __source__

```python
painting['src']
```

__Saving__ it on computer

```python
f = open('my_painting_image.jpg','wb')
```
